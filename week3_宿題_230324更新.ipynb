## 問題文
以下のサイトにあるデータ（ワインの品質）が分析対象になります。

http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

このデータを読み込み、カラムの`volatile acidity`について$n$等分（$n$はデータ数を越えず、分位数に同一の値が存在しない自然数。データ数が$n$で割り切れるとは限らず、この場合の処理は`Pandas.qcut`の処理に準じます。）にグループ分けします。次にそれぞれのグループのデータのうち、カラムの`quality`の値が`5`であるものについて、それらの`alcohol`の平均値を算出してください。さらに、ここで算出した各グループの`alcohol`の平均値の中で、1番小さい値を返り値とするような関数を作成してください。

```python
# common
import numpy as np
import pandas as pd
from pandas import DataFrame
# webからデータを取得したり、zipファイルを扱うためのライブラリ
import requests, zipfile
from io import StringIO
import io
# URL
url_winequality_data = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
```

```python
# データをurlから取得する
r = requests.get(url_winequality_data, stream=True).content
# データをDataFrameとして読み込む
df_winequality_data = pd.read_csv(io.BytesIO(r), sep=";")
df_winequality_data.head()
```

```python
df_winequality_data.columns
```

```python
# columnsの名前にスペースがあるときは、df.columnsの名前　が使えないので、df["columnsの名前"]とする
df_winequality_data["volatile_acidity_qcut"]= pd.qcut(df_winequality_data["volatile acidity"], 2)
```

```python
# 'quality'が5である行を選択し、それらを'volatile_acidity_qcut'でグループ化。その中で'alcohol'の平均を計算
mean_alcohol_quality5_per_bin = df_winequality_data[df_winequality_data['quality'] == 5].groupby('volatile_acidity_qcut')['alcohol'].mean()
```

```python
print(mean_alcohol_quality5_per_bin)
print(min(mean_alcohol_quality5_per_bin))
```

```python
# working place. everything
def homework(url_winequality_data, n):
    # データをurlから取得する
    r = requests.get(url_winequality_data, stream=True).content
    # データをDataFrameとして読み込む
    df_winequality_data=pd.read_csv(io.BytesIO(r), sep=";")
    
    # qcut関数でn分割し、"volatile_acidity_qcut"という名前の列として新たにdfに加える
    # columnsの名前にスペースがあるときは、df.columnsの名前　が使えないので、df["columnsの名前"]とする
    df_winequality_data["volatile_acidity_qcut"] = pd.qcut(df_winequality_data["volatile acidity"], n)
    
    # 'quality'が5である行を選択し、それらを'volatile_acidity_qcut'でグループ化。その中で'alcohol'の平均を計算
    mean_alcohol_quality5_per_bin = df_winequality_data[df_winequality_data['quality'] == 5].groupby('volatile_acidity_qcut')['alcohol'].mean()
    
    # 各グループ内におけるalcoholの平均値の最小値を返す
    my_result = min(mean_alcohol_quality5_per_bin)
    
    return my_result
```
