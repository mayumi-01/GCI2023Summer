```python
# common
import numpy as np
import pandas as pd
from sklearn import linear_model
import requests, zipfile
import io
from sklearn.model_selection import train_test_split

# URL
path_winequality_data = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
```

```python
# データをurlから取得する
r = requests.get(path_winequality_data, stream=True).content
# データをDataFrameとして読み込む
df_winequality_data = pd.read_csv(io.BytesIO(r), sep=";")
df_winequality_data.head()
```

```python
X = df_winequality_data["pH"]
X.head()
```

```python
# working place. everything
def homework(path_winequality_data, X_column, Y_column):
    # データをDataFrameとして読み込む
    df_winequality_data = pd.read_csv(path_winequality_data, sep=";")
    # 目的変数yにY_columnを指定、説明変数XにX_columnを指定
    # model.fitさせる際には、2次元配列である必要がある。そのため、reshape(-1, 1)で行数そのままで列数を1にする。
    X = df_winequality_data[str(X_column)].values.reshape(-1, 1)
    y = df_winequality_data[str(Y_column)]
    # モデル学習
    model = linear_model.LinearRegression()
    model.fit(X, y)
    # 決定係数を表示
    my_result = model.score(X, y)
    return my_result
```
